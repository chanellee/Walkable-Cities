```python
city_pop = pd.read_csv("sub-est2021_all.csv",encoding='latin-1')
state_codes = pd.read_csv("state_codes.csv")

city_pop = city_pop[["NAME", "STNAME","POPESTIMATE2021"]]
remove_states = [not i for i in city_pop["NAME"] == city_pop["STNAME"]]
remove_county = [not i.lower().__contains__("county") for i in city_pop["NAME"]] 
city_pop = city_pop[remove_states and remove_county]
city_pop = city_pop[city_pop["POPESTIMATE2021"]>=50000]
city_pop = city_pop.set_axis(['city', 'state', 'pop_estimate_2021'], axis='columns', copy = False)
state_codes = state_codes[["state","code"]].set_axis(['state', 'state_code'], axis='columns', copy = False)
city_pop = pd.merge(city_pop,state_codes, on = "state", how = "left")
city_pop["city_state"] = city_pop["city"] + ", " + city_pop["state"]
city_pop
```

```python
city_land = pd.read_csv("2020_Gaz_place_national.txt",sep = "\t",encoding='latin-1')
state_codes = pd.read_csv("state_codes.csv")

test = city_land[["USPS","NAME","ALAND"]].set_axis(['state_code', 'city', 'land_area'], axis='columns', copy = False)
city_land = pd.merge(test,state_codes[["state","code"]], left_on = "state_code", right_on = "code", how = "left")
city_land = city_land.dropna()
del city_land["code"]
city_land["city_state"] = city_land["city"] + ", " + city_land["state"]
city_land
```

```python
walking_data = pd.merge(city_pop,city_land, on = "city_state", how = "inner")
walking_data.set_axis([i.replace("_x","") for i in list(walking_data)], axis='columns', copy = False)
[i.__contains__("_y") for i in list(walking_data)]
```
